## Задача 4. Обмен половинок

На вход программе подается строка текста. Напишите программу, которая раздлит ее на две равные части, переставит их местами и выведет на экран.

### Пример

**Ввод:**

```
Привет, Андрей!
```

**Вывод:**

```
Андрей!Привет,
```

````python
s = input()
````

````python
# Ввод строки текста
s = input("Введите строку текста: ")

# Определение длины строки
length = len(s)

# Разделение строки на две равные части
middle = length // 2
first_half = s[:middle]
second_half = s[middle:]

# Перестановка и вывод на экран
result = second_half + first_half
print("Результат:", result)

````

## Задача 5. Таракан

Убежавший студент из задания 2 позвал на помощь с переписью вещей своего друга, который тараканов просто игнорирует. Необходимо вывести на экран вещи из списка студента, а если среди вещей появится '`таракан`' — на экран его не выводить, но цикл продолжить.

Примечение: необходимо предлагать ввод вещей с клавиатуры.  

### Пример

**Ввод:**

```
учебник
тетрадь
грязная кружка
таракан
степлер
носки
```

**Вывод:**

```
учебник
тетрадь
грязная кружка
степлер
носки
```

````python
things = []


while True:
    thing = input("Введите вещь студента (для завершения введите 'готово'): ")
    if thing.lower() == 'готово':
        break
    things.append(thing)

for thing in things:
    if thing.lower() == 'таракан':
        continue
    print(thing)


````

## Задача 6. Инвентаризация в художественной мастерской

Владелец художественной мастерской ежегодно делает заказ на приобретение следующих материалов: Кисть,
Краски,
Разбавитель,
Шпатель,
Холст,
Палитра.

Помогите ему подсчитать количество приобретенных материалов за 3 года, если дано количество по каждому году.

Результат вывести в формате:
Количество приобретенных материалов вида `material` за 3 года составило `summa` шт.



***Примечание: Решите задачу двумя способами без использования zip и с использованием.***


### Пример

**Ввод:**

```
[60, 15, 5, 2, 5, 15]
[45, 8, 3, 2, 3, 13]
[10, 8, 3, 2, 0, 10]

```


```
Количество приоберетенной посуды вида "Кисть" за 3 года составило 115 шт.
Количество приоберетенной посуды вида "Краски" за 3 года составило 31 шт.
Количество приоберетенной посуды вида "Разбавитель" за 3 года составило 11 шт.
Количество приоберетенной посуды вида "Шпатель" за 3 года составило 6 шт.
Количество приоберетенной посуды вида "Холст" за 3 года составило 8 шт.
Количество приоберетенной посуды вида "Палитра" за 3 года составило 38 шт.
```



````python
#без использования
materials_year1 = {'Кисть': 10, 'Краски': 20, 'Разбавитель': 15, 'Шпатель': 8, 'Холст': 25, 'Палитра': 5}
materials_year2 = {'Кисть': 15, 'Краски': 25, 'Разбавитель': 18, 'Шпатель': 10, 'Холст': 30, 'Палитра': 6}
materials_year3 = {'Кисть': 12, 'Краски': 22, 'Разбавитель': 20, 'Шпатель': 9, 'Холст': 28, 'Палитра': 7}

total_materials = {}
for material in materials_year1:
    total_materials[material] = materials_year1[material] + materials_year2[material] + materials_year3[material]

for material, total in total_materials.items():
    print(f"Количество приобретенных материалов вида {material} за 3 года составило {total} шт.")

#с использованием
materials_year1 = {'Кисть': 10, 'Краски': 20, 'Разбавитель': 15, 'Шпатель': 8, 'Холст': 25, 'Палитра': 5}
materials_year2 = {'Кисть': 15, 'Краски': 25, 'Разбавитель': 18, 'Шпатель': 10, 'Холст': 30, 'Палитра': 6}
materials_year3 = {'Кисть': 12, 'Краски': 22, 'Разбавитель': 20, 'Шпатель': 9, 'Холст': 28, 'Палитра': 7}

total_materials = {}
for (material1, quantity1), (material2, quantity2), (material3, quantity3) in zip(materials_year1.items(), materials_year2.items(), materials_year3.items()):
    total_materials[material1] = quantity1 + quantity2 + quantity3

for material, total in total_materials.items():
    print(f"Количество приобретенных материалов вида {material} за 3 года составило {total} шт.")
````

## Задача 6. Сотрудники

В программе задан пустой словарь workers.

Программа принимает на вход 3 параметра для каждого из сотрудников: имя, должность и зарплату.

Напишите функцию, которая принимает на вход имя, должность и зарплату сотрудника, и добавляет его в словарь по следующему правилу: ключом словаря являются имена работников, а значением — вложенный словарь, содержащий информацию о должности и зарплате.

1) Вызовите вашу функцию для переданных параметров, а затем выведите полученный словарь на экран.

2) Задан непустой словарь workers с информацией о сотрудниках предприятия (как в .1). Простая печать словаря при помощи функции print() довольно тяжело читается пользователем. Усовершенствуйте вывод информации из словаря: используя метод .items() и цикл for выведите построчно информацию о каждом рсотрудника
3) Представим себе, что Алексей Трегубов (информация о нем есть в словаре) только что прошел аккредитацию на мидл-разработчика.

Программа принимает на вход 2 переменные: строковую — с названием новой позиции сотрудника, и числовую — с новым значением его зарплаты.

Замените позицию и зарплату сотрудника в словаре на переданные в соответствующих переменных значения. Выведите на экран информацию о данном сотруднике, обратившись к словарю по имени сотрудника.

### Пример

**Для п. 1.**

**Ввод:**

```
Василий Петрович
инженер
100
```

**Вывод:**

```
{'Василий Петрович': {'position': 'инженер', 'salary': 100}}
```


---


**Для п. 2.**

**Ввод:**

```
Василий Петрович
инженер
100
```

**Вывод:**

```
Алексей Трегубов {'должность': 'программист', 'позиция': 'младший', 'зарплата': 120000}
```


---


**Для п. 3.**

**Ввод:**

```
мидл
190000
```

**Вывод:**

```
{'должность': 'программист', 'позиция': 'мидл', 'зарплата': 190000}
```


````python
# Пустой словарь сотрудников
workers = {}

# Функция для добавления сотрудника в словарь
def add_worker(name, position, salary):
    workers[name] = {'position': position, 'salary': salary}

# Вызов функции для добавления сотрудников
add_worker('Иванов', 'Менеджер', 50000)
add_worker('Петров', 'Разработчик', 60000)
add_worker('Сидоров', 'Тестировщик', 55000)

# Вывод полученного словаря
print(workers)

# Перебираем словарь сотрудников и выводим информацию о каждом
for name, info in workers.items():
    print(f"Сотрудник: {name}")
    print(f"Должность: {info['position']}")
    print(f"Зарплата: {info['salary']} руб.\n")


# Изменение информации о сотруднике
workers['Трегубов'] = {'position': 'Мидл-разработчик', 'salary': 70000}

# Вывод обновленной информации о сотруднике
print(f"Информация о сотруднике Алексее Трегубове:")
print(f"Должность: {workers['Трегубов']['position']}")
print(f"Зарплата: {workers['Трегубов']['salary']} руб.")


````
